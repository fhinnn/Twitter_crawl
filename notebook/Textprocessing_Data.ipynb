## Fungi tokenize

```python
def clean_text(text):
    try:
        cleaned_text = re.sub(r'[^\w\s]', '', text)
        return cleaned_text
    except Exception as e:
        print(f"An error occurred during text cleaning: {e}")
        return ""

def tokenize_text(text):
    # Check if the input is a string, if not, return an empty list
    if not isinstance(text, str):
        return []

    try:
        tokens = word_tokenize(text)
        tokens_with_quotes = ['"' + token + '"' for token in tokens]
        return tokens_with_quotes
    except Exception as e:
        print(f"An error occurred during tokenization: {e}")
        return []
```

```python
df_pemilu['Tweets'] = df_pemilu['Tweets'].apply(clean_text)
df_pemilu['Tweets'] = df_pemilu['Tweets'].apply(tokenize_text)
```

## Stemming text

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory # steeming
```

```python
nltk.download('stopwords') # download data stopwords bahasa Indonesia dari NLTK
```

```python
# filter kata penghubung, tambahan ke dalam list
additional_stopwords = ["dan", "atau", "juga", "serta", "sebagai", "yang", "yg", "banget", "gak"]
stopwords.extend(additional_stopwords)
```

```python
def stem_text(tokens):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    filtered_tokens = [token for token in stemmed_tokens if token.lower() not in stopwords]
    return filtered_tokens
```

```python
df_pemilu_subset = df_pemilu.head(20).copy()
# df_pemilu_subset['Tweets'] = df_pemilu_subset['Tweets'].apply(lambda x: nltk.word_tokenize(x))
df_pemilu_subset['Tweets'] = df_pemilu_subset['Tweets'].apply(stem_text)
```
