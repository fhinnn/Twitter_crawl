Read Dataset from Google drive

```python
url = 'https://drive.google.com/uc?id=1pdaeJ0ThUJcuFmrUzeN_U8ZCApG1yBmT'
df_pemilu = pd.read_csv(url)
df_pemilu.head(20)
```

Case Folding pada Text Dataset

```python
df_pemilu['Tweets'] = df_pemilu['Tweets'].apply(lambda x: x.lower())
df_pemilu.head(20)
```

Fungi tokenize

```python
def clean_text(text):
    cleaned_text = re.sub(r'[^\w\s]', '', text)
    return cleaned_text

def tokenize_text(text):
    tokens = word_tokenize(text)
    tokens_with_quotes = ['"' + token + '"' for token in tokens]
    return tokens_with_quotes
```

```python
df_pemilu['Tweets'] = df_pemilu['Tweets'].apply(clean_text)
df_pemilu['Tweets'] = df_pemilu['Tweets'].apply(tokenize_text)
df_pemilu.head(20)
```

## Stemming text

```python
def stem_text(tokens):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    return stemmed_tokens
```

```python
df_pemilu_subset = df_pemilu.head(20).copy()
df_pemilu_subset['Tweets'] = df_pemilu_subset['Tweets'].apply(stem_text)
```
