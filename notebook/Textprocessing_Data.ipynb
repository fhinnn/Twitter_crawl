Case Folding pada Text Dataset

```python
df_pemilu['Tweets'] = df_pemilu['Tweets'].apply(lambda x: x.lower())
df_pemilu.head(15)
```

Fungi tokenize

```python
def clean_text(text):
    cleaned_text = re.sub(r'[^\w\s]', '', text)
    return cleaned_text

def tokenize_text(text):
    tokens = word_tokenize(text)
    tokens_with_quotes = ['"' + token + '"' for token in tokens]
    return tokens_with_quotes
```

```python
df_pemilu['Tweets'] = df_pemilu['Tweets'].apply(clean_text)
df_pemilu['Tweets'] = df_pemilu['Tweets'].apply(tokenize_text)
df_pemilu.head(50)
```
